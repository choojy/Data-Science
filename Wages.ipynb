# 3) Continuing from above, how would you examine and clean outlier?

```python
# boxplot of the lwage
sns.boxplot(x=df["lwage"])
```

```python
def calculate_outliers(column):
    
    # Calculating 25th and 75th percentile.
    percentile25 = np.percentile(column,25)
    print("perctile 25",percentile25)
    percentile75 = np.percentile(column,75)
    print("percentile 75",percentile75)
    
    diff = percentile75 - percentile25
    
    lowerlimit = percentile25 - diff * 1.5
    upperlimit = percentile75 + diff * 1.5
    
    column[column<lowerlimit] = np.nan
    column[column>upperlimit] = np.nan
    
    return column
```

```python
#create a new data frame by dropping the first column
df_final = df.drop('Unnamed: 0', axis=1).copy()
df_final.head()
```

```python
#check outlier and remove the value in lwage
df_final['lwage'] = calculate_outliers(df_final['lwage'])
```

```python
#change label to number for modelling
df_final['bluecol'] = df_final['bluecol'].map({'yes': 1, 'no': 0})
df_final['south'] = df_final['south'].map({'yes': 1, 'no': 0})
df_final['smsa'] = df_final['smsa'].map({'yes': 1, 'no': 0})
df_final['married'] = df_final['married'].map({'yes': 1, 'no': 0})
df_final['union'] = df_final['union'].map({'yes': 1, 'no': 0})
df_final['black'] = df_final['black'].map({'yes': 1, 'no': 0})
df_final['sex'] = df_final['sex'].map({'male': 1, 'female': 0})
df_final.head()
```

```python
# Imputation with KNN missing value after remove the outlier
from sklearn.impute import KNNImputer
knn_imputer = KNNImputer(n_neighbors=5)
columns = df_final.columns
df_final = knn_imputer.fit_transform(df_final)
df_final = pd.DataFrame(df_final,columns=columns)
```

```python
#check boxplot after remove outlier and imputation
sns.boxplot(x=df_final["lwage"])
```

# 5) Build a model from data, share with us the insights and results.

```python
y = df_final['lwage'].copy()
x = df_final.drop('lwage', axis=1).copy()
```

```python
# importing train_test_split from sklearn
from sklearn.model_selection import train_test_split
# splitting the data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
# importing module of Linear Regression
from sklearn.linear_model import LinearRegression

LR = LinearRegression()
res = LR.fit(x_train,y_train)
y_prediction =  LR.predict(x_test)
```

```python
# importing r2_score module
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error

# predicting the accuracy score
score=r2_score(y_test,y_prediction)
print('r2 score is ',score)
print('mean_sqrd_error is==',mean_squared_error(y_test,y_prediction))
print('root_mean_squared error is==',np.sqrt(mean_squared_error(y_test,y_prediction)))
```

```python
# retrieve coefficient of the model
cdf = pd.DataFrame(LR.coef_, x_train.columns, columns=['Coefficients'])
cdf.sort_values(by=['Coefficients'], ascending=False)
```

```python
# importance of coefficients
coefs = pd.DataFrame(
    LR.coef_* x_train.std(axis=0),
    columns=["Coefficient importance"],
    index=x_train.columns,
).sort_values(by=['Coefficient importance'], ascending=True)
coefs.plot(kind="barh", figsize=(9, 7))
plt.xlabel("Coefficient values corrected by the feature's std. dev.")
plt.axvline(x=0, color=".5")
plt.subplots_adjust(left=0.3)
```

Findings:

1.   The RMSE of the model is 0.334 which is very small. Thus the model is good to predict logarithm of wage
2.   The features that have positive impact on the wages are years in education, sex, work experience, resides in a standard metropolitan statistical area, individual's wage set by a union contract, maritial status, weeks worked and works in a manufacturing industry
3.   black individual, resides in south and blue collar have negative impact on the wages
4.   The most important feature is years in education. Each year increase in education, there is an increase in logarithm of wage by 0.0478.
5.   The second important feature is sex. A male worker has algorithm of wage of 0.325 higher compare to female




# 6) Find a way to classify this dataset into 3 homogenous groups.

```python
# the data will be divided into 3 clusters using KMeans Clustering
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# select variables for clustering
X = df_final.copy()

# standardize variables
scaler = StandardScaler()
X = scaler.fit_transform(X)

# perform k-means clustering
kmeans = KMeans(n_clusters=3, random_state=0).fit(X)

# add cluster labels to original data
df_final["cluster"] = kmeans.labels_

# plot the clusters
colors = ["red", "green", "blue"]
plt.figure(figsize=(12,8))
for i in range(3):
    cluster_data = df_final[df_final["cluster"] == i]
    plt.scatter(cluster_data["lwage"], cluster_data["ed"], color=colors[i], label=f"Cluster {i+1}")

plt.xlabel("Wage")
plt.ylabel("Years of Education")
plt.legend()
plt.show()
```
